```python
# 以下是抽帧代码
video = cv2.VideoCapture(r"E:\数据蛙\数据蛙视频转换器\Converted\1th-ship.mp4")
num = 0                      # 计数器
save_step = 10               # 间隔帧
while True:
    ret,frame = video.read()
    if not ret:
        break
    num += 1
    if num % save_step == 0:
        cv2.imwrite("./1th-images/" + str(num) +".jpg", frame)
```
